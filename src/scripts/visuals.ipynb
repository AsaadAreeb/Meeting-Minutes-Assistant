```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import matplotlib.patches as patches
from IPython.display import HTML

# Define the components of the sequence diagram
components = ["User", "Frontend/UI", "Bucket", "Speech-to-Text API", "Gemini API"]

# Define the sequence of interactions
sequence = [
    ("User", "Frontend/UI", "Initiates audio upload/recording"),
    ("Frontend/UI", "Bucket", "Uploads audio file"),
    ("Frontend/UI", "Speech-to-Text API", "Sends audio link"),
    ("Speech-to-Text API", "Bucket", "Fetches audio file"),
    ("Speech-to-Text API", "Frontend/UI", "Returns transcription"),
    ("Frontend/UI", "Gemini API", "Sends transcription"),
    ("Gemini API", "Frontend/UI", "Returns classified text"),
    ("Frontend/UI", "User", "Displays classified text"),
]

# Adjust the spacing between components
scaling_factor = 2  # Adjust this factor to increase/decrease spacing
x_positions = [i * scaling_factor for i in range(1, len(components) + 1)]

# Diagram settings
fig, ax = plt.subplots(figsize=(12, 6))  # Increase figure size for better spacing
ax.set_xlim(0, x_positions[-1] + scaling_factor)
ax.set_ylim(0, len(sequence) + 1)
ax.axis('off')

# Draw component lifelines
for x, component in zip(x_positions, components):
    ax.text(x, len(sequence) + 0.5, component, ha="center", fontsize=12, weight="bold")
    ax.plot([x, x], [0.5, len(sequence) + 0.2], linestyle="--", color="gray", alpha=0.5)

# Prepare annotations for the interactions
annotations = []
arrows = []
text_offset = 0.2  # Vertical offset for text placement above arrows
for i, (sender, receiver, message) in enumerate(sequence):
    sender_idx = components.index(sender)
    receiver_idx = components.index(receiver)
    sender_x = x_positions[sender_idx]
    receiver_x = x_positions[receiver_idx]
    # Add the text annotation above the arrow
    annotations.append(
        ax.text(
            (sender_x + receiver_x) / 2, len(sequence) - i + text_offset,
            message, ha="center", va="center", fontsize=10, alpha=0
        )
    )
    # Add the arrow
    arrow = patches.FancyArrow(
        sender_x, len(sequence) - i, receiver_x - sender_x, 0,
        width=0.05, length_includes_head=True, color="blue", alpha=0
    )
    ax.add_patch(arrow)
    arrows.append(arrow)

# Animation function
def animate(frame):
    if frame < len(sequence):
        annotations[frame].set_alpha(1)  # Show the message
        arrows[frame].set_alpha(1)  # Show the arrow

# Create animation
ani = FuncAnimation(
    fig, animate, frames=len(sequence), interval=1000, repeat=False
)

# Display animation in Jupyter Notebook
HTML(ani.to_jshtml())
```
